```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("C:/Users/MSI/OneDrive/Desktop/listings.csv")

# Data Cleaning: Fill missing values for price and remove NaN rows in key columns
data['price'] = data['price'].fillna(0)
data.dropna(subset=['neighbourhood', 'price', 'latitude', 'longitude'], inplace=True)

# Group by neighborhood to calculate listings count and average price
neighborhood_summary = data.groupby('neighbourhood').agg(
    listings_count=('id', 'count'),
    avg_price=('price', 'mean')
).reset_index()

# Sort by listings count to identify high-concentration neighborhoods
neighborhood_summary = neighborhood_summary.sort_values(by='listings_count', ascending=False)

# Plotting: Bar chart for high-concentration neighborhoods and average pricing
plt.figure(figsize=(12, 6))
sns.barplot(
    x='listings_count',
    y='neighbourhood',
    data=neighborhood_summary.head(10),
    palette='viridis'
)
plt.title('Top 10 Neighborhoods by Listing Count')
plt.xlabel('Number of Listings')
plt.ylabel('Neighborhood')
plt.show()

# Scatter plot for average pricing trends
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x='listings_count',
    y='avg_price',
    data=neighborhood_summary,
    hue='neighbourhood',
    legend=False,
    palette='tab10'
)
plt.title('Neighborhood Listings vs. Average Price')
plt.xlabel('Number of Listings')
plt.ylabel('Average Price (USD)')
plt.show()
```
